해결했으니 함수로 만들어서 모든 데이터의 날짜를 datetime 형태로 만들어버리자

```python
import re
def make_date(df):
    
    
    # 정규 표현식으로 년월일 날리기
    temp = []
    for date in df['날짜']:
        numbers = re.findall("\d+", date)
        temp.append(numbers)
        
    # 날리고 남은 숫자들을 합치는데 중간에 - 를 넣으면서 합치기 다 만들어진 친구들은 str임
    datetime=[]
    for date in temp:
        datetime.append("-".join(date))
    
    df['datetime'] = datetime
    df.index = df['datetime']
    del df['날짜']
    del df['datetime']
    return df
```

```python
df_list = [Gold,Silver,DollarIndex,US_10_year_rate,SPX,DJIA,USD_KRW]

for i in df_list:
    make_date(i)
```
